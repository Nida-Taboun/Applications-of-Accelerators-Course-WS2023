Q11- Write a function in Python code that multiples 2 (two) 3x3 Matrices (ndarray) and returns the result.
Call this function using two randomly generated 3 by 3 matrices and print the results.

```python
import numpy as np
```

```python
# function
def multiply_matrices(matrix1, matrix2):
        return np.dot(matrix1, matrix2)
```

```python
matrix_a = np.random.rand(3,3)
```

```python
matrix_b = np.random.rand(3,3)
```

```python
#call the function
result = multiply_matrices(matrix_a, matrix_b)
```

```python
#print random matrix A
print(matrix_a)
```

```python
#print random matrix B
print(matrix_b)
```

```python
#prent the result
print(result)
```
