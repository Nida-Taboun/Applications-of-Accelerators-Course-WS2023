Q7- Write a super small python class for a particle, don't forget the init, by giving Z and A. this class should have a method which gives the number of neutron back. Instan-tiate this class with an example.

```python
#Inputs
print("            Example to calculate number of neutrons            ")
Z = input("Enter the proton number (z)")
A = input("Enter the atomic mass (A):")
#---------------------------------------
class Neutron_Calculator:
    def __init__(self, Z, A):
        self.Z = Z
        self.A = A
#---------------------------------------
# Calculate the number of neutrons in the particle.
#---------------------------------------
    def Calculate_number_of_neutrons(self):
        N = int(self.A) - int(self.Z)
        print("The number of neutrons in this particle is = %d." % N)
#---------------------------------------       
Example = Neutron_Calculator(Z,A)
Example.Calculate_number_of_neutrons()
```
